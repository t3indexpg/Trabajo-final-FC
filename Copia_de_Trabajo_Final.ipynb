### **Ejercicio**

La empresa minera ABC tiene un proyecto de explotación de dos minas de hierro. La mina Paracatu produce cada día 4 toneladas de hierro de alta calidad, 5 toneladas de calidad media y 9 de baja calidad. La mina Bejang produce cada día 1 toneladas de hierro de alta calidad, 4 toneladas de calidad media y 15 de baja calidad.
La compañía necesita al menos 70 toneladas de hierro de alta calidad, 150 toneladas de calidad media y 350 de baja calidad. El coste diario de la mina Paracatu es de 1.980USD y la mina Bejang 1.530USD en cada mina ¿Cuántos días debe trabajar cada mina para que el coste sea mínimo?


```python
# Data
df = pd.DataFrame({' ' : ['Mina Paracatu', 'Mina Bejang',''],
        'Días' : ['x', 'y',''],
        'Alta Calidad' : ['4x', 'y','70'],
        'Calidad Media' : ['5x', '4y','150'],
        'Baja Calidad' : ['9x', '15y','350'],
        'Coste Diario' : ['$1.980x', '$1.530y','']})
df
```

### Optimización

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.optimize import linprog
```

```python
# Parametros de la función objetivo
MinaH = np.array([1980,1530])

# Restricciones de desigualdad
M1 = np.array([[4,1],[5,4],[9,15]])
R1 = np.array([70,150,350])

# Solución del problema
RTA = linprog(-MinaH, A_ub=M1, b_ub=R1, method='revised simplex')
RTA
```

RTA: 50.670 es el coste mínimo que debe pagar la empresa si en la mina Paracatu trabajan 14 días y en la mina Bejang 15 días para obtener los rendimientos esperados que son: al menos 70 toneladas de hierro de alta calidad, 150 toneladas de calidad media y 350 de baja calidad.

### **Técnica de simulación de Monte Carlo**

La empresa planea introducirse directamente como una comercializadora. Para estimar la utilidad neta del primer año se tienen los siguientes componentes: • Volumen de ventas (unidades) • Precio por unidad (precio de venta) • Costo unitario • Costos fijos
Los costos fijos (como gastos generales, arriendo, servicios, etc.) se estiman en $100.000 USD a razón de los costos por el uso de las minas y se asume que son determinísticos. Las demás variables involucran cierta incertidumbre: i) el volumen de ventas (en unidades) puede cambiar en un rango amplio que depende del grado de aceptación del producto y de la competencia; ii) el precio de venta por unidad dependerá de las acciones que emprenda la competencia y; iii) los costos unitarios también variarán según las interacciones del mercado. Además, se asume que todas estas variables son independientes.

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def betapert(x,y,z,*,size=1,lamb=4):
  a=z-x
  alpha=1+lamb*(y-x)/a
  beta=1+lamb*(z-y)/a
  return x+np.random.beta(alpha,beta,size=size)*a
```

```python
cantidad = betapert(500,650,700,size=1000)
cantidad
```

```python
precio = betapert(1600,2000,2100,size=1000)
precio
```

```python
ingreso=precio*cantidad
```

```python
costos=np.random.uniform(1530,1980,1000)
costos
```

```python
costosfijos = 100000
```

```python
utilidad =(cantidad*precio)-(cantidad*costos)-costosfijos
```

```python
pd.Series(utilidad).hist()
utilidad.mean()
```
